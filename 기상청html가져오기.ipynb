```python
import requests                  # 웹 페이지의 HTML을 가져오는 모듈
from bs4 import BeautifulSoup    # HTML을 파싱하는 모듈
 
# 웹 페이지를 가져온 뒤 BeautifulSoup 객체로 만듦
response = requests.get('https://www.weather.go.kr/w/obs-climate/land/city-obs.do')
soup = BeautifulSoup(response.content, 'html.parser')
import csv
 
table = soup.find('table', { 'class': 'table-col' })    # <table class="table_develop3">을 찾음
data = []                            # 데이터를 저장할 리스트 생성
for tr in table.find_all('tr'):      
    tds = list(tr.find_all('td'))    
                                     
    for td in tds:                   
        if td.find('a'):             # <td> 안에 <a> 태그가 있으면(지점인지 확인)
            point = td.find('a').text    # <a> 태그 안에서 지점(도시)를 가져옴
            nowweather = tds[1].text
            km = tds[2].text
            cloud = tds[3].text
            cloudw = tds[4].text
            temperature = tds[5].text    # <td> 태그 리스트의 인덱스 5- 기온
            drop = tds[6].text
            feel = tds[7].text
            rain = tds[8].text
            humidity = tds[9].text    
            windpos = tds[10].text
            windspeed = tds[11].text
            pressure = tds[12].text   
            data.append([point, nowweather, km, cloud, cloudw,temperature, 
                         drop, feel, rain, humidity, windpos, windspeed, pressure])    # data 리스트에 지점, 기온, 습도를 추가
 
#print(data)    # data 확인 
import csv

# Define the column names
column_names = ['(지점)이름', '현재 일기', '시정km', '운량1/10', '중하운량', '현재 기온', '이슬점 온도', '체감 온도', '일강수 (mm)', '습도%', '풍향', '풍속 m/s']

with open('weather.csv', 'w', newline='', encoding = 'utf-16') as file:
    writer = csv.writer(file)
    # Write the column names
    writer.writerow(column_names)
    writer.writerows(data)

f = open("weather.csv", 'r', encoding = 'utf-16')
line = f.read()
print(line)
f.close()
```
