```python
import gdown
gdown.download('https://bit.ly/3RhoNho', 'ns_202104.csv', quiet = False)
```

```python
import pandas as pd
ns_df = pd.read_csv('ns_202104.csv', low_memory = False)
ns_book = ns_df.loc[:,'번호':'등록일자'] #번호부터 등록일자까지 표 자르기 
ns_book.head()
```

중간 행을 제외한 표 만들기 

```python
#Index 출력
print(ns_df.columns)
#해당하는 원소 찾기 (== 혹은 != 사용)
#ns_df.columns == 'Unnamed: 13'
ns_df.columns != '대출건수'
```

```python
selected_columns = ns_df.columns != '부가기호'

ns_book = ns_df.loc[:, selected_columns]
ns_book.head()
```

```python
ns_df.drop(['부가기호','Unnamed: 13','세트 ISBN'], axis = 1, inplace = True)
ns_df.head()
```

NaN이 하나 이상 있는 행이나 열 삭제 - dropna() 

```python
ns_book = ns_df.dropna(axis =1)
ns_book.head()
```

모든 값이 NaN인 값 삭제 = dropna(how = 'all')

```python
ns_book = ns_df.dropna(axis = 0, how = 'all')  #axis = 0: 기본, 행 제거(가로) axis = 1: 열(세로) 제거 
ns_book.head(10)
```

행 삭제하기 

```python
ns_book2 = ns_book.drop([0,1])   #0부터 1까지 행 삭제 
#= ns_book2 = ns_book[2:]
ns_book2.head()
```

중복된 행 찾기 

```python
print(sum(ns_book.duplicated()))
#설명 - 파이썬의 sum()함수를 이용하면 True를 1로 인식하기 때문에 반환값이 1 이상이다. 
print()
print(sum(ns_book.duplicated(subset=['도서명','저자','ISBN'])))
```

```python
dup_rows = ns_book.duplicated(subset=['도서명','저자','ISBN'],keep=False)
ns_book3 = ns_book[dup_rows]
ns_book3.head()
```

그룹별로 모으기- 같은 도서는 하나로 합치고, 대출건수와 도서 수 등도 합친다.

```python
count_df = ns_book[['도서명','저자','ISBN','권','대출건수']]
group_df = count_df.groupby(by = ['도서명','저자','ISBN','권'],dropna = False)  #원래 groupby()메서드는 NaN이 하나라도 있는 행을 삭제하는데, 
#dropna = False는 삭제하지 않도록 한다.
loan_count = group_df.sum()  #groupby()함수: []안의 데이터가 같다면 같은 개체로 보고, 그 외의 데이터는 합친다. 
loan_count.head()
```

```python
dup_rows = ns_book.duplicated(subset=['도서명','저자','ISBN','권'])

unique_rows = ~dup_rows   #boolean 반전
ns_book3 = ns_book[unique_rows].copy()
sum(ns_book3.duplicated(subset=['도서명','저자','ISBN','권']))  #0으로 나오면 중복된 행 없음음
```

```python
ns_book3.set_index(['도서명','저자','ISBN','권'], inplace = True)  #지정한 열을 인덱스로 만들어줌 
ns_book3.head()
```

```python
ns_book3.update(loan_count)
ns_book3.head()
```

```python
ns_book4 = ns_book3.reset_index()
ns_book4.head()
```

```python
print("데이터 정리 이전 대출건수 100회 이상인 도서 수: ")
print(sum(ns_book['대출건수']>100))
print()
print("데이터 정리 이후 대출건수 100회 이상인 도서 수: ")
sum(ns_book4['대출건수']>100)
```

```python
ns_book4 = ns_book4[ns_book.columns]
ns_book4.to_csv('ns_book4.csv',index = False)
```
