```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
fmri = sns.load_dataset('fmri')
sns.lineplot(data = fmri, x= 'timepoint',y = 'signal', hue = 'event',style ='region')
_=plt.show()
```

```python
sns.boxplot(x = 'event',y = 'signal',hue = 'region',data = fmri)
plt.grid = False
_=plt.show()
```

```python
iris = sns.load_dataset('iris')
iris.boxplot(by = 'species')
plt.tight_layout(pad = 3, h_pad = 1)
_=plt.show()
```

```python
sns.boxplot(x = 'sepal_length',y = 'species',data =iris)
_=plt.show()
```
