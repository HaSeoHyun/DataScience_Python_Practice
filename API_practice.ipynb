```python
d = {"name":"hongongData", "me":"하서현"}
print(d['name'])
import json
#json.dupmps: 파이썬 객체를 JSON 형식의 텍스트로 
d_str = json.dumps(d,ensure_ascii = False) #ensure_ascii: 한글 출력
print(d_str)
print(type(d_str))  #데이터 타입 확인 
```

```python
d2 = json.loads(d_str)  #json.loads()함수: JSON 문자열을 파이썬 객체로 변환 
print(d2['name'])
print(type(d2))
```

```python
x_str = '''
<book>  
  <name>혼자 공부하는 데이터 분석</name>   
  <author>박해선</author>
  <year>2022</year> 
</book>
'''

'''(주석)
<book>  #루트, 부모   
  <name>혼자 공부하는 데이터 분석</name>   #<name>: 시작 태그, </name>: 종료태그 
  <author>박해선</author>
  <year>2022</year>  # name, author, year: 자식 
</book>  #종료 태그 
'''

import xml.etree.ElementTree as et
book = et.fromstring(x_str)
print(type(book))
print(book.tag)
```

```python
book_childs = list(book)  #자식 엘리먼트 확인 
print(book_childs)  #자식 엘리먼트 : name, author, year 
name = book.findtext('name')  #findtext()함수로 객체 찾기 
author = book.findtext('author')
year = book.findtext('year')
print(name)
print(author)
print(year)
```

```python
import requests  #웹 서버로부터 받은 JSON문자열을 파이썬 객체로 변환하여 반환 
url = "http://data4library.kr/api/loanItemSrch?format=json&startDt=2021-04-01&endDt=2021-04-30&age=20&authKey=56477804e4a1b1d7aed549ec135ccac0b9a372a3c6ea23fc6fbf7b5a3f833ed2"
r = requests.get(url)
data = r.json()

#print(data)
books = []  #books변수에 도서 정보가 담긴 딕셔너리 리스트 만들기 
for d in data['response']['docs']:
  books.append(d['doc'])
```

```python
books_df = pd.DataFrame(books)
books_df
```

```python
books_df.to_json('20s_best_book.json')  #만든 목록을 json파일에 저장 
```

```python
import requests  #웹 서버로부터 받은 JSON문자열을 파이썬 객체로 변환하여 반환 
url = "http://data4library.kr/api/loanItemSrch?format=json&startDt=2022-04-01&endDt=2022-04-30&age=20&authKey=56477804e4a1b1d7aed549ec135ccac0b9a372a3c6ea23fc6fbf7b5a3f833ed2"
r = requests.get(url)
data = r.json()   #json()함수: json 문자열을 파이썬 객체로 변환하여 반환한다. 
#print(data)

books = []  #books변수에 도서 정보가 담긴 딕셔너리 리스트 만들기 
for d in data['response']['docs']:
  books.append(d['doc'])
books_df = pd.DataFrame(books)
books_df
#url을 손보면 날짜나 찾는 기준 등을 변경하여 사용할 수 있다. 
```
