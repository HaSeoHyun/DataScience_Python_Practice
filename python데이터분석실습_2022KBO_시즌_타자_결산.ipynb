```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/kbo_2022_hitter_rank.csv',
                 encoding = 'EUC-KR', low_memory = False)
```

Data Cleaning (Fixing Team Names)

```python
df_new = df.replace({'(.*)삼(.*)':'삼성','(.*)키(.*)': r'키움','(.*)K(.*)': r'KIA','(.*)L(.*)': r'LG','(.*)S(.*)': r'SSG','(.*)두(.*)': r'두산',
            '(.*)한(.*)': r'한화','(.*)롯(.*)': r'롯데','(.*)N(.*)': r'NC'}, regex=True)
df_new.head()
```

KT 선수들이 KIA로 표기된 오류가 발생 -> 정정 

```python
df_new.loc[16, '팀'] = 'KT'
df_new.loc[22, '팀'] = 'KT'

df_new.head(30)
```

```python
df_new.describe()  #top 30 선수들의 평균, max, min, 25,50,75백분위
```

데이터 시각화하기
1. 산점도를 이용하여 타석과 득점관의 관계 시각화 

```python
import matplotlib.pyplot as plt
plt.scatter(df_new['타석'], df_new['득점'])
plt.show()
```

위 산점도의 x축은 타석, y축은 득점이다. 대체로 비례하는 모습을 보인다. 

-> Linear Regression model로 사용가능하다.

2. 히스토그램을 이용하여 안타 수 시각화 

```python
plt.hist(df_new['안타'])
plt.show()
```

**머신러닝으로 타석에 따른 득점 예측하기 **

```python
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(df_new)
#print(len(train_set), len(test_set))
```

```python
X_train = train_set[['타석']]
y_train = train_set['득점']

from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)

X_test = test_set[['타석']]
y_test = test_set['득점']

lr.score(X_test, y_test)  #결정 계수(coefficient of determination)
```
