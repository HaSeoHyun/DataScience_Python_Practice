```python
#(c) adding column names

import pandas as pd

df = pd.read_csv('iris.data', header=None)
df.columns = ['petal_width','petal_length', 'sepal_width' ,'sepal_length' ,'species'] 
#df.head()

#(d) how many species of flower?
# Print the result
#value_counts = df['species'].value_counts()
#print(value_counts)

#Iris-setosa        50
#Iris-versicolor    50
#Iris-virginica     50
#Name: species, dtype: int64
```

```python
#2. correcting information
row_35 = df.iloc[34]  # 35번 row를 row_35로 정의
#print(row_35)
row_38 = df.iloc[37]
#print(row_38)


df.at[34, 'sepal_length'] = 0.2
# Print
print("changed 35th row:")
print(row_35)
print()

df.at[37, 'petal_length'] = 3.6
df.at[37, 'sepal_width'] = 1.4
print("changed 38th row:")
print(row_38)
```

```python
#3. Add two additional features
df['Petal_ratio'] = df['petal_length'] / df['petal_width']
df['Sepal_ratio'] = df['sepal_length'] / df['sepal_width']

# Print the updated DataFrame
print(df.head())

#4
#df.to_csv('iris_new.csv', index=False)
```

```python
#5
statistics = df.groupby('species')[['Petal_ratio', 'Sepal_ratio']].agg(['mean', 'median', 'min', 'max', 'std'])

# Print the statistics
print(statistics)
'''
                Petal_ratio                                          
                       mean    median       min       max       std   
species                                                               
Iris-setosa        0.684248  0.683502  0.511111  0.788462  0.051871   
Iris-versicolor    0.467680  0.462687  0.354839  0.566667  0.046829   
Iris-virginica     0.453396  0.460928  0.337662  0.548387  0.047015   

                Sepal_ratio                                          
                       mean    median       min       max       std  
species                                                              
Iris-setosa        0.167868  0.142857  0.066667  0.375000  0.065789  
Iris-versicolor    0.311106  0.308608  0.243902  0.375000  0.029213  
Iris-virginica     0.366739  0.375000  0.250000  0.470588  0.050232  
'''
```

```python
#II.Plotting Iris dataset 

#1.선그래프 
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# CSV 파일 불러오기
df = pd.read_csv('iris_new.csv')
species = df['species'].unique()

# 그래프 그리기
plt.figure(figsize=(8, 6))  # 플롯 크기 설정

# 종(species)별로 그래프 그리기
for sp in species:
    subset = df[df['species'] == sp]
    Petal_ratio = subset['Petal_ratio']
    row_number = subset.index
    plt.plot(row_number, Petal_ratio, label=sp)

plt.xlabel('Row Number')
plt.ylabel('Petal_ratio')
plt.title('Petal ratio Plot')
plt.legend()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# CSV 파일 불러오기
df = pd.read_csv('iris_new.csv')

# 그래프 그리기
plt.figure(figsize=(8, 6))  # 플롯 크기 설정

# KDE plot 그리기
sns.kdeplot(data=df, x="Petal_ratio", hue="species", fill=True)

plt.xlabel('Petal_ratio')
plt.ylabel('Density')
plt.title('KDE Plot of Petal ratio by Species')
plt.show()
```

```python
#II-2. Scatter matrix

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the CSV file into a pandas DataFrame
df = pd.read_csv('iris_new.csv')

# Define the four features for the scatter matrix plot
features = ['petal_width', 'petal_length', 'sepal_width', 'sepal_length']

# Set the color palette for different species
color_palette = {'Iris-setosa': 'orange', 'Iris-versicolor': 'green', 'Iris-virginica': 'blue'}

# Create a pairplot using seaborn
sns.set(style="ticks")
sns.pairplot(df, vars=features, hue='species', palette=color_palette)

# Adjust the plot layout
plt.tight_layout()

# Display the scatter matrix plot
plt.show()
```
